```python
%%sql
SELECT
  DATE_TRUNC('month',orderdate):: date AS order_month,
  SUM (s.quantity * s.netprice * s.exchangerate) AS net_revenue,
  COUNT (DISTINCT customerkey) AS total_unique_customers
FROM
  sales s
GROUP BY order_month

```

```python
%%sql
SELECT
  orderdate,
  TO_CHAR(orderdate,'MM-YYYY') AS YEAR
FROM
  sales
ORDER BY RANDOM()
LIMIT 10
```

```python
%%sql
SELECT
  TO_CHAR(orderdate,'MM-YYYY') AS order_month_year ,
  SUM (s.quantity * s.netprice * s.exchangerate) AS net_revenue,
  COUNT(DISTINCT (customerkey)) AS total_unique_customers
FROM
  sales s
GROUP BY order_month_year
```

```python
%%sql
SELECT
  orderdate,
  DATE_PART('year', orderdate) AS YEAR,
  DATE_PART('month', orderdate) AS MONTH,
  DATE_PART('day', orderdate) AS day
FROM
  sales s
ORDER BY RANDOM()
LIMIT 10



```

```python
%%sql

SELECT
  orderdate,
  EXTRACT(YEAR FROM orderdate) AS extract_year,
  EXTRACT(MONTH FROM orderdate) AS extract_month,
  EXTRACT(DAY FROM orderdate) AS extract_day
FROM
  sales
ORDER BY RANDOM()
LIMIT 10
```

```python
%%sql
SELECT
  EXTRACT(YEAR FROM orderdate) AS order_year,
  EXTRACT(MONTH FROM orderdate) AS order_month,
  SUM(s.quantity * s.netprice * s.exchangerate) AS net_revenue
FROM
  sales s
GROUP BY
  order_year, order_month
ORDER BY
  order_year, order_month
```

```python
%%sql

SELECT
  CURRENT_DATE,
  EXTRACT(YEAR FROM orderdate) AS order_year,
  EXTRACT(YEAR FROM CURRENT_DATE) AS current_year,
  EXTRACT(YEAR FROM CURRENT_DATE) -5 AS minus_five,
  s.orderdate,
  p.categoryname,
  SUM(s.quantity * s.netprice * s.exchangerate) AS net_revenue
FROM sales s
LEFT JOIN product p ON s.productkey = p.productkey
WHERE
 orderdate >= (CURRENT_DATE - INTERVAL '5 YEARS')
GROUP BY
  s.orderdate,
  p.categoryname
ORDER BY
  s.orderdate,
  p.categoryname
```

```python
%%sql

SELECT
  CURRENT_DATE,
  s.orderdate,
  p.categoryname,
  SUM(s.quantity * s.netprice * s.exchangerate) AS net_revenue
FROM sales s
LEFT JOIN product p ON s.productkey = p.productkey
WHERE
  orderdate >= (CURRENT_DATE - INTERVAL '5 YEARS')
GROUP BY
  s.orderdate,
  p.categoryname
ORDER BY
  s.orderdate,
  p.categoryname
```

```python
%%sql
SELECT
  AGE('2020-12-01','2020-01-01') AS interval,
  EXTRACT(DAY FROM AGE('2020-12-08', '2020-01-01')) -5 AS day_interval

```

```python
%%sql

SELECT
  DATE_PART('YEAR', s.orderdate) AS order_year,
  ROUND(AVG(EXTRACT(DAY FROM AGE(s.deliverydate,s.orderdate))), 2)  AS avg_processing_time,
  CAST(SUM(quantity * netprice* exchangerate) AS INTEGER) AS net_revenue
  # EXTRACT HAS TO BE USED TO TURN AGE INTO INTEGER SO AVG CAN BE USED WITH IT
FROM
  sales s
WHERE
  orderdate>= CURRENT_DATE - INTERVAL '5 YEARS'
GROUP BY
  order_year
ORDER BY
  order_year
```
